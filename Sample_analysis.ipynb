```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

sentimentAnalyser = SentimentIntensityAnalyzer()
```

```python
import pandas as pd
pd.options.display.max_colwidth = 400
```

```python
import pandas as pd
tweets_sample = pd.read_excel('/content/drive/MyDrive/Python final project/Sentiment analysis/Sample IT.xlsx')
tweets_sample.head()
```

```python
#Calculating sentiment analysis
```

```python
def calculate_sentiment(text):
    scores = sentimentAnalyser.polarity_scores(text)
    compound_score = scores['compound']
    return compound_score
```

```python
tweets_sample['sentiment_score'] = tweets_sample['Tweet'].apply(calculate_sentiment)
```

```python
#Calculating score for top 10 entries
```

```python
tweets_sample.sort_values(by='sentiment_score', ascending=False)[:10]
```
