```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER so we can use it later
sentimentAnalyser = SentimentIntensityAnalyzer()
```

```python
import pandas as pd
tweets=pd.read_csv('/content/drive/MyDrive/Python final project/Sentiment analysis/islamophobiatweets.csv')
tweets.head()
```

```python
tweets
```

```python
def calculate_sentiment(text):
    scores = sentimentAnalyser.polarity_scores(text)
    compound_score = scores['compound']
    return compound_score
```

```python
tweets['sentiment_score'] = tweets['Tweet'].apply(calculate_sentiment)
```

```python
tweets.sort_values(by='sentiment_score', ascending=False)[:10]
```

```python
tweets.sort_values(by='sentiment_score', ascending=True)[:10]
```
