```python
import requests
import pprint
```

```python
url = 'https://newsapi.org/v2/everything?language=en&q=terror&apiKey=e2b3ff0bcdd2414a95009021f1a062f1'
```

```python
response = requests.get(url)
```

```python
response_json = response.json()
pprint.pprint(response_json)
```

```python
#Looking at just the headlines of the articles
```

```python
for i in response_json['articles']:
    print(i['title'])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text_combined = ''
```

```python
for i in response_json['articles']:
    text_combined += i['title'] + ' '
print(text_combined[0:300])
```

```python
wordcloud = WordCloud(max_font_size=40).generate(text_combined)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
